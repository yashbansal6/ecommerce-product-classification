## Scraping all products from a URL

```python
from tqdm import tqdm
import time
import pandas as pd
import selenium
from selenium.webdriver import Chrome
from selenium.common.exceptions import NoSuchElementException
```

```python
base_url = "https://mamaearth.in/product-category/beauty"
path_to_driver = "C:/WebDriver/bin/chromedriver.exe"
```

```python
driver = Chrome(executable_path=path_to_driver)
driver.get(base_url)
```

```python
driver
```

```python
driver.current_url
```

```python
items = driver.find_elements_by_class_name("uniquewhite")
```

```python
# get the links of all products present on the page
objects = []
for _ in items:
    obj = _.find_element_by_tag_name('a').get_property('href')
    objects.append(obj)
```

```python
#using XPath to find all the required elements
product_name=[]
product_link=[]
product_rating=[]
product_review=[]
product_mrp=[]
product_pack_size=[]
product_discount=[]
product_ingredient=[]

for obj in tqdm(objects):
    
    #time.sleep(1)
    driver.get(obj)
    
    try:
        name = driver.find_element_by_xpath('//*[@id="__next"]/div[2]/div[1]/div[2]/div[2]/div/h1/div').text
    except NoSuchElementException:
        name = driver.find_element_by_xpath('//*[@id="__next"]/div[2]/div[1]/div[3]/div[2]/div/h1/div').text
    except NoSuchElementException:
        name = 'None'
        
    link = obj
    
    try:
        rating = driver.find_element_by_xpath('//*[@id="__next"]/div[2]/div[1]/div[2]/div[2]/div/div[1]/a[1]/div/div/div[1]').text
    except NoSuchElementException:
        rating = 'None'
    if(rating=='None'):
        try:
            rating = driver.find_element_by_xpath('//*[@id="__next"]/div[2]/div[1]/div[3]/div[2]/div/div[1]/a[1]/div/div/div[1]').text
        except NoSuchElementException:
            rating = 'None'
    
    try:
        review = driver.find_element_by_xpath('//*[@id="__next"]/div[2]/div[1]/div[2]/div[2]/div/div[1]/a[2]').text
    except NoSuchElementException:
        review = 'None'
        
    try:
        mrp = driver.find_element_by_xpath('//*[@id="__next"]/div[2]/div[1]/div[2]/div[2]/div/div[2]/div[1]').text
    except NoSuchElementException:
        mrp = 'None'
        
    #pack_size = driver.find_element_by_xpath('').text
    
    try:
        discount = driver.find_element_by_xpath('//*[@id="__next"]/div[2]/div[1]/div[2]/div[2]/div/div[2]/div[3]').text
    except NoSuchElementException:
        discount = 'None' 

 
    try:
        ingredient = driver.find_element_by_xpath('//*[@id="__next"]/div[2]/div[1]/div[3]/div[2]/div/ul/li[1]/span/strong').text
    except NoSuchElementException:
        ingredient = 'None'
            
    product_name.append(name)
    product_link.append(link)
    product_rating.append(rating)
    product_review.append(review)
    product_mrp.append(mrp)
    #product_pack_size.append(pack_size)
    product_discount.append(discount)
    product_ingredient.append(ingredient)
    
    #print(name, link, rating, review, mrp, discount, review)
```

```python
dict = {'product_name':product_name,
    'product_link':product_link,
    'product_rating':product_rating,
    'product_review':product_review,
    'product_mrp':product_mrp,
    #product_pack_size.append(pack_size),
    'product_discount':product_discount,
    'product_ingredient':product_ingredient} 

df = pd.DataFrame(dict)
df
```

```python
df.to_csv("scraped_products.csv")
```

```python
dicts ={hair_shampoo_url : 'hair-shampoo',
        hair_conditioner_url : 'hair-conditioner',
        hair_oil_url : 'hair-oil',
        hair_mask_url : 'hair-mask',
        hair_serum_url : 'hair-serum',
        face_wash_url : 'face-wash',
        face_mask_url : 'face-mask',
        face_cream_url : 'face-cream',
        face_serum_url : 'face-serum',
        face_scrub_url : 'face-scrub',
        face_toner_url : 'face-toner',
        face_gel_url : 'face-gel',
        face_sheet_mask_url : 'face-sheet-mask',
        body_butter_url : 'body-butter',
        body_wash_url : 'body-wash',
        body_lotion_url : 'body-lotion',
        body_scrub_url : 'body-scrub',
        hand_cream_url : 'hand-cream'}
```

```python
dataset = []
path_to_driver = "C:/WebDriver/bin/chromedriver.exe"
driver = Chrome(executable_path=path_to_driver)

# getting names of all products in given category

for url, product_class in dicts.items():
    
    driver.get(url)
    items = driver.find_elements_by_class_name("uniquewhite")
    objects = []
    
    for _ in items:
        obj = _.find_element_by_tag_name('a').get_property('href')
        objects.append(obj)
    
    for obj in tqdm(objects):
        driver.get(obj)
        try:
            name = driver.find_element_by_xpath('//*[@id="__next"]/div[2]/div[1]/div[2]/div[2]/div/h1/div').text
        except NoSuchElementException:
            name = driver.find_element_by_xpath('//*[@id="__next"]/div[2]/div[1]/div[3]/div[2]/div/h1/div').text
        except NoSuchElementException:
            name = 'None'

        data = {'product_name':name, 'product_class':product_class}
        dataset.append(data)
```

```python
df = pd.DataFrame(dataset)

# as the dataset is continuous in accordance to the categories, we shuffle it for proper distribution
df2 = df.sample(frac=1).reset_index(drop=True)
df2.to_csv("classes_data.csv")
df2
```

## ML Model for Product Category classification

```python
import pandas as pd
import re
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn import model_selection, preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from xgboost import XGBClassifier
```

```python
classes_df = pd.read_csv('classes_data.csv')
```

```python
classes_df
```

```python
classes_df.drop(['Unnamed: 0'],axis=1, inplace=True)
classes_df
```

```python
product_classes = classes_df['product_class'].unique()
n_classes = len(product_classes)
print(n_classes)
print(product_classes)
```

```python
# preprocessing the text for better learning in the model
for index, row in classes_df.iterrows():
    row['product_name'] = row['product_name'].lower()# making all words lower case           
    row['product_name'] = re.sub(r'\W+', ' ', row['product_name']) #removing non-aplhanumeric values
```

```python
classes_df
```

```python
tf = TfidfVectorizer(token_pattern=r'\w+', max_features=5000)

X = tf.fit_transform(classes_df['product_name']).toarray()
y = classes_df['product_class']

encoder = preprocessing.LabelEncoder()
y = encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify=classes_df['product_class'])
```

```python
param={'objective': 'multi:softprob'}
param['eval_metric'] = ['mlog_loss']
xgb_model = XGBClassifier(param, use_label_encoder=False, learning_rate=1e-3, reg_lambda=1, tree_method='exact',
                      subsample=1, sampling_method='gradient_based')

xgb_model.fit(X_train, y_train)

y_pred = xgb_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

## ML Model for Product Ingredient classification

```python
all_products = pd.read_csv('scraped_products.csv')
all_products
```

```python
all_products.drop(['Unnamed: 0'],axis=1, inplace=True)
sorted(all_products['product_ingredient'].unique())
# we can see that there are many duplicates as well as non-english entries
```

```python
# preprocessing for removing duplicates and better learning in the model
for index, row in all_products.iterrows():
    
    row['product_name'] = row['product_name'].lower() #lower case
    row['product_name'] = re.sub(r'\W+', ' ', row['product_name']) #removing non-alphanumeric chars
    
    row['product_ingredient'] = row['product_ingredient'].lower() #lower case
    row['product_ingredient'] = re.sub(r':', '', row['product_ingredient']) #removing colons
    row['product_ingredient'] = re.sub(r' \Z', '', row['product_ingredient'])
    row['product_ingredient'] = re.sub('[^a-zA-Z0-9 ]','',row['product_ingredient'])#removing non-english entries
    row['product_ingredient'] = re.sub(r' \Z', '', row['product_ingredient']) #removing trailing whitespaces

#remove rows with following values
all_products= all_products[all_products.product_ingredient != 'none']
all_products= all_products[all_products.product_ingredient != '']
all_products= all_products[all_products.product_ingredient != ' ']
all_products= all_products[all_products.product_ingredient != '  ']
```

```python
#classes have reduced as compared to earlier
sorted(all_products['product_ingredient'].unique())
```

```python
tf = TfidfVectorizer(token_pattern=r'\w+', max_features=5000)

X = tf.fit_transform(all_products['product_name']).toarray()
y = all_products['product_ingredient']

enc = preprocessing.LabelEncoder()
y = enc.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
param={'objective': 'multi:softprob'}
param['eval_metric'] = ['mlog_loss']
xgb_model = XGBClassifier(param, use_label_encoder=True, learning_rate=1, reg_lambda=1, tree_method='exact',
                      subsample=1, sampling_method='gradient_based')

xgb_model.fit(X_train, y_train)

y_pred = xgb_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
